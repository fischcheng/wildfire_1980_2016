### From now on the data is in Pandas dataframe.
* Filter major fire >1000 acre
* Filter West US <103W
* Groupby YEAR_ : count and sum(TOTALACRES)
* Calculate the duration > time difference between STARTDATED and CONTRDATED
* Calculate the length of the wildfire season > the first start of the year to the last controled of the year. 
* Difficult to define the ending of the wildfire season > max set to 365
* ISSUE: only the data from Forest Service, not included some suburb area?

```python
dff.dtypes
```

```python
dff=pd.DataFrame(df[['CAUSE','SPECCAUSE','SIZECLASSN','YEAR_','STARTDATED','CONTRDATED','OUTDATED','STATE','TOTALACRES','DLONGITUDE']])

# Clean data
dff['STARTDATED']=pd.to_datetime(dff['STARTDATED'])
dff['CONTRDATED']=pd.to_datetime(dff['CONTRDATED'])
dff['OUTDATED']=pd.to_datetime(dff['OUTDATED'])
dff['LENGTH']=(dff['CONTRDATED']-dff['STARTDATED'])

dff['YEAR']=pd.to_numeric(dff['YEAR_'])
dff=dff.drop('YEAR_',axis=1)

WEST_US_major_fire=dff[(dff['SIZECLASSN']>5) & (dff['DLONGITUDE']<-103)]
```

```python
def season_length(year):
    startdate=WEST_US_major_fire[WEST_US_major_fire['YEAR']==year].sort_values('STARTDATED')['STARTDATED'].iloc[0]
    if year == 2015:
        enddate=WEST_US_major_fire[WEST_US_major_fire['YEAR']==year].sort_values('OUTDATED',na_position='first')['OUTDATED'].iloc[-1]
    else:
        enddate=WEST_US_major_fire[WEST_US_major_fire['YEAR']==year].sort_values('CONTRDATED',na_position='first')['CONTRDATED'].iloc[-1]
    #print(startdate)
    #print(enddate)
    return(enddate-startdate)

season_window=[]
for year in range(1980,2017):
    season_window.append(season_length(year).days)   

#season_window # some years doesn't have major fires >> so probably not a good idea to use >1000 as threshold
```

```python
event_count=WEST_US_major_fire.groupby('YEAR')['STARTDATED'].count()
total_acre=WEST_US_major_fire.groupby('YEAR').sum()['TOTALACRES']
#year=pd.date_range(start='1980', end='2017', freq='A')
season=pd.Series(season_window,index=total_acre.index)
season[season>365]=365
annual_df=pd.DataFrame({'event_count': event_count, 'total_acre': total_acre,'season_length':season})
```

```python
import seaborn as sns
sns.set_style("white")
season[season>365]=365
ax=season.plot(color='orange')
plt.title('Average Wildfire Length from 1980-2016')
ax.set_xlabel('Year')
ax.set_ylabel('Days')
sns.despine()
```

```python
from numpy.polynomial.polynomial import polyval,polyfit

regressions=polyfit(total_acre.index,total_acre,1)
rline=polyval(total_acre.index,regressions)
plt.bar(total_acre.index,total_acre,color='orange')
#plt.plot(total_acre.index,rline,'red')
sns.despine()
```

```python
regressions=polyfit(total_acre.index,event_count,1)
rline=polyval(total_acre.index,regressions)

plt.bar(total_acre.index,event_count,color='orange')
#plt.plot(total_acre.index,rline,'red')
sns.despine()
```

```python
regressions=polyfit(total_acre.index,season,1)
rline=polyval(total_acre.index,regressions)
plt.bar(season.index,season,color='orange')
#plt.plot(total_acre.index,rline,'red')
sns.despine()
```
